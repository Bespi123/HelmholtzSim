## **Simulate Helmholtz coils with Circular shape in a single Axis**

Import Libraries

```python
import time
import numpy as np
import pandas as pd
import parallel_test4 as test_3
import plotMagneticField as hplot
```

Simulation Settings

```python
# Initialize coil parameters
X_coil = test_3.CoilParameters(2, 0.5864538565842049, 30)

# Current coil simulation
I = 1
# Rotation matriz
Ax = np.eye(3) ## eye rotation matrix

# Simulation settings
parallel_coils = 2
batch_Size = 120
grid_length_size = 0.01 #Space among evaluation points
num_seg = 100           #Numer of segments 
```

Set Spires geometry

```python
spire1_x_c, spire2_x_c = test_3.square_spires(Ax, X_coil.h, X_coil.a, num_seg)
```

```python
spire1_x_c, spire2_x_c = test_3.circular_spires(Ax, X_coil.h, X_coil.a, num_seg)
```

```python
##Spawn spires
spire1_x_c, spire2_x_c = test_3.star_spires(Ax, X_coil.h, X_coil.a, num_seg, 5)
f0 = None
f0 = hplot.plot_spires(f0, spire1_x_c, spire2_x_c, color='black', row=None, col=None)

# Plot the square spires
f0.show()
```

```python
##Spawn spires
spire1_x_c, spire2_x_c = test_3.polygonal_spires(Ax, X_coil.h, X_coil.a, num_seg, 6)
f0 = None
f0 = hplot.plot_spires(f0, spire1_x_c, spire2_x_c, color='black', row=None, col=None)

# Plot the square spires
f0.show()
```

Set Simulation grid

```python
# Generar grid
X, Y, Z = test_3.generate_range(X_coil.a*3/4, grid_length_size)

hplot.plot_grid(X, Y, Z, f0)
```

## Run Simulation ##

```python
#Run simulations
start_time = time.time() #Count start time
x_coil_results_c = test_3.coil_simulation_1d_sequential(X, Y, Z, X_coil, I, spire1_x_c, spire2_x_c, parallel_coils, batch_Size)
end_time = time.time()   #Mark ending time

# Calcular la norma del campo magnético B = sqrt(Bx^2 + By^2 + Bz^2)
x_coil_results_c["B_norm"] = np.sqrt(x_coil_results_c["Bx"]**2 + x_coil_results_c["By"]**2 + x_coil_results_c["Bz"]**2)

#Calculate and show the simulation time
execution_time = end_time - start_time
print(f'Simulation finished in {execution_time/60} minutes...')

# Save results in a CSV file
output_file = 'x_coil_results_c.csv'
x_coil_results_s.to_csv(output_file, index=False)
```
